```python
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()
```

```python
df['title'] = df['title'].str.replace('"', '')
```

```python
df['releaseYear'] = pd.to_numeric(df['releaseYear'], errors='coerce').astype('Int64')
```

```python
df['imdbAverageRating'] = pd.to_numeric(df['imdbAverageRating'], errors='coerce')
```

```python
df['imdbNumVotes'] = pd.to_numeric(df['imdbNumVotes'], errors='coerce').astype('Int64')
```

```python
df['genres'] = df['genres'].str.replace('"', '').str.split(', ')
```

```python
df['availableCountries'] = df['availableCountries'].fillna('').str.split(', ')
```
